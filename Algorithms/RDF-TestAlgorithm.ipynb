```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Define the path to the folder with good and bad sunset images
data_path = '/Users/d/Desktop/Sunsets/SunsetPics/bad sunsets'

# Load the images and labels
data = []
labels = []
for filename in os.listdir(data_path):
    if filename.endswith('.jpg'):
        img = cv2.imread(os.path.join(data_path, filename))
        img = cv2.resize(img, (64, 64))  # Resize images to a fixed size
        label = 1 if 'good' in filename else 0
        data.append(img)
        labels.append(label)

# Convert data to a numpy array and scale pixel values to [0, 1]
data = np.array(data, dtype=np.float32) / 255.0
labels = np.array(labels)

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.2, random_state=42)

# Train a random forest classifier with 100 trees
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(train_data.reshape(len(train_data), -1), train_labels)

# Evaluate the accuracy of the classifier on the testing set
accuracy = clf.score(test_data.reshape(len(test_data), -1), test_labels)
print(f"Accuracy: {accuracy:.2f}")

# Use the trained classifier to predict the labels of new images
new_data_path = '/Users/d/Desktop/Sunsets/SunsetPics/bad sunsets'
for filename in os.listdir(new_data_path):
    if filename.endswith('.jpg'):
        img = cv2.imread(os.path.join(new_data_path, filename))
        img = cv2.resize(img, (64, 64))
        data = np.array([img], dtype=np.float32) / 255.0
        prediction = clf.predict(data.reshape(len(data), -1))
        if prediction[0] == 1:
            print(f"{filename} is a good sunset.")
        else:
            print(f"{filename} is a bad sunset.")
```
